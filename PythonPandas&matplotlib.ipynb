```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", None],
    "Age": [25, None, 30, 22]
})

df["Age"].fillna(df["Age"].mean(), inplace=True)

# Convert Name column to uppercase
df["Name"] = df["Name"].str.upper()
print(df)
```

```python
import pandas as pd
import numpy as np
#sample data
exam_data={
    'name':['Anastasia','Dima','Katherina','james','Emily','Michael','Mathew','Kevin','Laura','Jonas'],
    'score':[12.5,9,16.5,np.nan,9,20,14.5,np.nan,8,19],
    'attempts':[1,3,2,3,2,3,1,1,2,1],
    'qualify':['yes','no','yes','no','no','yes','yes','no','no','yes']
}
labels =['a','b','c','d','e','f','g','h','i','j']

#a)create and display dataframe

df =pd.DataFrame(exam_data,index=labels)
print("Initial DataFrame:\n",df)

#b)Change 'James' to 'suresh' in name column

df.loc[df['name'] == 'James','name'] = 'Suresh'
print("\nAfter renaming James to Suresh:\n",df)

#c)Insert a new column in DataFrame

df['remarks'] = ['Good','Average','Excellent','Poor','Average','Excellent','Good','Poor','Weak','Excellent']
print("\nAfter inserting new column 'remarks':\n",df)

#d)Get list of column headers

col_list = df.columns.tolist()
print("\nList of column headers:",col_list)
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,label="Line Plot",color="blue",linewidth=1,marker="o")
plt.title("Simple Line Plot for 1")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.legend()
plt.grid()
plt.show()
```

```python
categories = ["A","B","C","D"]
values=[4,7,2,9]
plt.bar(categories,values)
plt.title("Simple Bar Plot")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
import numpy as np
categories=["A","B","C","D"]
values1=[4,7,2,9]
values2=[5,6,3,8]

x=np.arange(len(categories))
width=0.35
plt.bar(x-width/2,values1,width,label="Group 1",color="blue")
plt.bar(x+width/2,values2,width,label="Group 2",color="red")
plt.xticks(x,categories)
plt.title("Grouped Bar Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

students=["Ashok","Babu","Chiru","Dayal","Emma"]
math_scores=[85,70,90,60,75]
science_scores=[80,65,95,70,85]

plt.plot(students,math_scores,marker='o',label="Math")
plt.plot(students,science_scores,marker='s',label="Science")
plt.title("Students Performance")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.show()


plt.scatter(math_scores,science_scores,color="red",marker="*",s=50)
plt.title("Math vs Science Scores")
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.show()

plt.bar(students,math_scores,color='blue',alpha=0.6,label="Math")
plt.bar(students,science_scores,color='orange',alpha=0.6,bottom=math_scores,label="Science")
plt.title("Stacked Bar Chart of Scores")
plt.legend()
plt.show()
```

```python
fig,axes = plt.subplots(1,4,figsize=(15,5))
axes[0].plot(students,math_scores,color='blue',marker='o')
axes[0].set_title("Line Plot")

axes[1].scatter(math_scores,science_scores,color='red')
axes[1].set_title("Scatter Plot")

axes[2].bar(students,science_scores,color='green')
axes[2].set_title("Bar Plot")

axes[3].barh(students,science_scores,color='green')
axes[3].set_title("Bar Plot")


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

dates = pd.date_range(start="2023-01-05",periods=10,freq="D")
values = np.random.randint(50,100,size=10)

plt.figure(figsize=(10,5))
plt.plot(dates,values,marker='o',linestyle='--',color='red')

plt.title("Time-Series Data")
plt.xlabel("Date")
plt.ylabel("Value")

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gcf().autofmt_xdate()

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x*2 + y*2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x,y,z,cmap = 'viridis')

ax.set_title("3D surface plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```
